```python
# unzip
## 前提：本ディレクトリに「地上気象観測(normal_surface.zip)」が配置されていること
## ダウンロード元：https://www.data.jma.go.jp/obd/stats/data/mdrr/normal/index.html
```

```python
# zipファイル内のファイル一覧を取得
import zipfile
zf = zipfile.ZipFile('./normal_surface.zip')
print(zf.namelist())
zf.close()
```

```python
# zipファイルから1ファイルだけ展開
import zipfile
zf = zipfile.ZipFile('./normal_surface.zip')
# カレントディレクトリにnormal_surfaceディレクトリが生成される
zf.extract('normal_surface/regional_average/daily/nml_sfc_ra_d_00000.csv')
zf.close()
```

```python
# zipファイルから全ファイル展開
import zipfile
zf = zipfile.ZipFile('./normal_surface.zip')
# カレントディレクトリにnormal_surfaceディレクトリが生成される
zf.extractall()
zf.close()
```

```python
# zipファイルにファイルを追加
import pathlib
import zipfile

newtxt_1 = pathlib.Path('./newtxt_1.txt')
newtxt_2 = pathlib.Path('./newtxt_2.txt')
newtxt_3 = pathlib.Path('./newtxt_3.txt')

newtxt_1.touch()
newtxt_2.touch()
newtxt_3.touch()

newzip = zipfile.ZipFile('newzip.zip', mode='w')
newzip.write(newtxt_1)
newzip.write(newtxt_2)
newzip.write(newtxt_3)
newzip.close()

newzip = zipfile.ZipFile('newzip.zip', mode='r')
print(newzip.namelist())
newzip.close()
```
